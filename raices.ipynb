```python
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
def polinomio(x):
    return x**5 - 6*x**4 + 2*x**3 + 20*x**2 - 27*x + 10
def derivada(x):
    return 5*x**4 - 24*x**3 + 6*x**2 + 40*x - 27
puntos_criticos = fsolve(derivada, np.linspace(-10, 10, 10))
puntos_iniciales = np.append(puntos_criticos, np.linspace(-100, 100, 100))
raices = fsolve(polinomio, puntos_iniciales)
raices_unicas = np.unique(np.round(raices, 6))
print("Raíces encontradas:", raices_unicas)
a=np.linspace(-100,100,100)
plt.plot(a,polinomio(a))

```

```python
import numpy as np
from scipy.optimize import root
def polinomio(x):
    return x**5 - 6*x**4 + 2*x**3 + 20*x**2 - 27*x + 10
puntos_iniciales = np.linspace(-10, 10, 20)  
soluciones = [root(polinomio, x0) for x0 in puntos_iniciales]
raices = np.unique(np.round([sol.x[0] for sol in soluciones if sol.success], 8))
print("Raíces encontradas:", raices)
```
